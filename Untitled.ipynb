Analysing trend of heart attack risk among genders

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count occurrences instead of summing
female_counts = file[(file['Gender'] == 'Female') & (file['Heart_Attack_Risk'] == 1)].shape[0]
male_counts = file[(file['Gender'] == 'Male') & (file['Heart_Attack_Risk'] == 1)].shape[0]

# Dictionary to store counts
dic = {'Female': female_counts, 'Male': male_counts}

# Convert to DataFrame
df = pd.DataFrame.from_dict(dic, orient='index', columns=['Count'])

# Plot using a bar chart
df.plot(kind='bar', legend=False, color=['pink', 'blue'])

# Add labels
plt.xlabel("Gender")
plt.ylabel("Heart Attack Cases")
plt.title("Heart Attack Risk by Gender")
plt.xticks(rotation=0)  # Keep labels readable

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count cases with and without heart attack history
ha_counts_with_ha = file[(file['Heart_Attack_History'] == 1) & (file['Heart_Attack_Risk'] == 1)].shape[0]
ha_not_ha = file[(file['Heart_Attack_History'] == 0) & (file['Heart_Attack_Risk'] == 1)].shape[0]

# Create dictionary
dic2 = {'No History': ha_not_ha, 'History': ha_counts_with_ha}

# Convert to DataFrame
df1 = pd.DataFrame.from_dict(dic2, orient='index', columns=['Count'])

# Plot bar chart
df1.plot(kind='bar', legend=False, color=['red', 'green'])

# Labels
plt.xlabel("Heart Attack History")
plt.ylabel("Number of Cases")
plt.title("Heart Attack Risk Based on History")
plt.xticks(rotation=0)  # Keep labels horizontal

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count heart attack risk based on air pollution exposure
ha_with_good_air = file[(file['Air_Pollution_Exposure'] == 1) & (file['Heart_Attack_Risk'] == 1)].shape[0]
ha_without_good_air = file[(file['Air_Pollution_Exposure'] == 0) & (file['Heart_Attack_Risk'] == 1)].shape[0]

# Create dictionary
dic3 = {
    'Inhales Unhealthy Air': ha_without_good_air,
    'Inhales Good Quality Air': ha_with_good_air
}

# Convert to DataFrame
df2 = pd.DataFrame.from_dict(dic3, orient='index', columns=['Count'])

# Plot bar chart
df2.plot(kind='bar', legend=False, color=['red', 'green'])

# Correct labels
plt.xlabel('Air Pollution Exposure')  # Fixed label
plt.ylabel('Risk of Heart Attack')
plt.title('Analysis of Heart Attack Risk Based on Air Pollution Exposure')
plt.xticks(rotation=90)  # Rotate x-axis labels by 90 degrees

# Show plot
plt.show()
```

```python
# Age group analysis
age_bins = pd.cut(file['Age'], bins=[0, 20, 30, 40, 50, 60, 70, 80], 
                  labels=['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80'])
risk_by_age = file.groupby(age_bins)['Heart_Attack_Risk'].mean()
plt.figure(figsize=(8, 5))
sns.barplot(x=risk_by_age.index, y=risk_by_age.values)
plt.title('Heart Attack Risk by Age Group')
plt.show()
```

```python

# Correlation Analysis
corr_vars = [col for col in file.columns if col not in ['State_Name', 'Gender']]
corr_matrix = file[corr_vars].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
